```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%config InlineBackend.figure_format = "retina"
```

```python
width = 17
height = 32
```

```python
grid = np.zeros((height, width))
grid[:, 4:8] = 1
grid[-6:, -8:] = 1
grid[-7:, -9] = 1

grid[18:27, 0] = 1
grid[11:28, 1] = 1
grid[4:30,  2] = 1
grid[0:31,  3] = 1

aspect_ratio = 2 * width / height
```

```python
fig, ax = plt.subplots()
# plt.matshow(grid, cmap="bone_r")
plt.imshow(grid, cmap="bone", extent=[0, width, 0, height])
plt.gca().invert_yaxis()
plt.gca().set_aspect(aspect_ratio, adjustable='box')

ax.set_xticks(np.arange(0, width, 1))
ax.set_yticks(np.arange(0, height, 1))
plt.grid(linewidth=0.2, color="tab:gray")
# Gridlines based on minor ticks
```
